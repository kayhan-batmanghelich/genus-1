```python
import pandas
import numpy as np


# read background SNPs List
bkgSNPTable = pandas.read_csv('/data/vision/polina/users/kayhan/Experiments/DPP/geneticResults/Data/bg.1kg.map.txt',sep='\t',header=None)
print "top 5 lines of the original table"
print bkgSNPTable.head(5)
print "========================"


# get only Chromosome 19
bkgSNPChr19Table = bkgSNPTable[ bkgSNPTable[:][1]=='chr19' ]
print "top 5 line of only chr19"
bkgSNPChr19Table.columns = ['SNP','Chr','physLoc','SNP_ext'] 
print bkgSNPChr19Table.head(5)
print "========================"

# read ADNI .bim file
adniSNPTable = pandas.read_csv('/data/vision/polina/users/kayhan/Experiments/DPP/geneticResults/Data/adni_bed2_ceu_flipped_clean.bim',sep='\t',header=None)
print "top 5 lines of the adni .bim file "
print adniSNPTable.head(5)
print "========================"

# extract chr 19
adniSNPChr19Table = adniSNPTable[ adniSNPTable[:][0]==19 ]
print "top 5 lines of the chr19 of the .bim file "
adniSNPChr19Table.columns = ['Chr','SNP','X','physLoc','A1','A2']
print adniSNPChr19Table.head(5)
print "========================"

```

```python
# find the background SNPs that are already in the list and
# remove the background SNPs that are already in the list


bkgSNP = np.array( bkgSNPChr19Table.SNP  )
bkgSNP = bkgSNP.squeeze()
commonIdx = [ (s in  np.array(adniSNPChr19Table.SNP) ) for s in bkgSNP]

commonSNP = bkgSNP[np.argwhere(commonIdx)]
commonSNP = commonSNP.squeeze()
#print commonSNP

diffIdx = [ not(s in  commonSNP ) for s in bkgSNP]
diffSNP = bkgSNP[np.argwhere(diffIdx)]
diffSNP = diffSNP.squeeze()
print diffSNP[0:10]

fid = open('/data/vision/polina/users/kayhan/Experiments/DPP/geneticResults/Data/chr19SNPExclusiveList.txt','wt')
for s in diffSNP:
    fid.write("%s\n"%s)

fid.close()    
```

```python
hapMapLD = pandas.read_csv('/data/vision/polina/users/kayhan/Experiments/DPP/geneticResults/Data/ld_chr19_CEU.txt',header=None,sep=' ')
```

```python
hapMapLD.columns = ['pos1','pos2','popCode','rs1','rs2','Dprime','Rsquare','LOD','fbin']
hapMapLD.head(5)
```

Make a union of the background SNPs and SNPs from Chr19 from ADNI dataset. Find only subset of SNPs that are simultaneously in "hapMap" and Chr19 of ADNI. 

```python
# make a union of the background SNPs and SNPs from Chr19 from ADNI dataset
allChr19SNPList = list(adniSNPChr19Table.SNP) + list(bkgSNPChr19Table.SNP)
allChr19SNPList = np.unique(allChr19SNPList)

# find rows of the tables in the set
hapMapLDAdni = hapMapLD[ hapMapLD['rs1'].isin(allChr19SNPList)] 
hapMapLDAdni = hapMapLDAdni[ hapMapLDAdni['rs2'].isin(allChr19SNPList)] 

```

```python

# find out what portion of background SNPs are not in the hapMAP
bkgSNPFoundIdx = np.logical_or( np.array( bkgSNPChr19Table['SNP'].isin(  hapMapLDAdni.rs1 ) )  ,  bkgSNPChr19Table['SNP'].isin(  hapMapLDAdni.rs2 ) )
bkgSNPFound = np.array( bkgSNPChr19Table.SNP[bkgSNPFoundIdx] ) 
print "background SNPs found in hapMAP : %d / %d  (%f)"%(sum(bkgSNPFoundIdx), len(bkgSNPChr19Table.SNP),  (sum(bkgSNPFoundIdx) + 0.0)/len(bkgSNPChr19Table.SNP) )
print "bkgSNP : ", bkgSNPFound[0:10] , "....."

# find out what portion of ADNI SNPs are not in the hapMAP
adniSNPFoundIdx = np.logical_or( np.array( adniSNPChr19Table['SNP'].isin(  hapMapLDAdni.rs1 ) )  ,  adniSNPChr19Table['SNP'].isin(  hapMapLDAdni.rs2 ) )
adniSNPFound = np.array( adniSNPChr19Table.SNP[adniSNPFoundIdx] )
print "ADNI SNPs found in hapMAP : %d / %d  (%f)"%(sum(adniSNPFoundIdx), len(adniSNPChr19Table.SNP),  (sum(adniSNPFoundIdx) + 0.0)/len(adniSNPChr19Table.SNP) )
print "adni snp :", adniSNPFound[0:10] , " ....."



# sort all ADNI SNPs based on physical position
adniPosFound = np.array(adniSNPChr19Table.physLoc[adniSNPChr19Table.SNP.isin(adniSNPFound)])
adniSNPFound = adniSNPFound[np.argsort(adniPosFound)]



# sort all background SNPs based on physical position
bkgPosFound = np.zeros((len(bkgSNPFound,)))
cnt = 0
for rs in bkgSNPFound:
    idx1 = np.where(np.array(hapMapLDAdni.rs1 == rs))
    idx2 = np.where(np.array(hapMapLDAdni.rs2 == rs))
    if not(idx1[0].size==0):
        bkgPosFound[cnt] = np.array(hapMapLDAdni[hapMapLDAdni.rs1 == rs]['pos1'])[0]
    elif not(idx2[0].size==0):
        bkgPosFound[cnt] = np.array(hapMapLDAdni[hapMapLDAdni.rs2 == rs]['pos2'])[0]
    else:
        print "nothing found for ", rs , cnt   
    if np.mod(cnt,100)==0:
        print "cnt : ", cnt
    cnt = cnt + 1 
bkgSNPFound = bkgSNPFound[ np.argsort(bkgPosFound) ]
```

```python
# playing with operators in pandas dataframe, making everything works as expected
idx1 = np.logical_and( np.array( hapMapLDAdni['rs1'].isin( bkgSNPFound )  ) , np.array( hapMapLDAdni['rs2'].isin( adniSNPFound )   ) )
idx2 = np.logical_and( np.array( hapMapLDAdni['rs1'].isin( adniSNPFound )  ) , np.array( hapMapLDAdni['rs2'].isin( bkgSNPFound )   ) )

idx1 = np.logical_and( np.array( hapMapLDAdni['rs1'] == bkgSNPFound[0] ) , np.array( hapMapLDAdni['rs2'].isin( adniSNPFound )   ) )
idx2 = np.logical_and( np.array( hapMapLDAdni['rs1'].isin( adniSNPFound )  ) , np.array( hapMapLDAdni['rs2'] == bkgSNPFound[0] ) )
print "rs :",  bkgSNPFound[0]
print "number non-zero LD (bkgSNP first):" , sum(idx1)
print "number non-zero LD (ADNI SNP first):" , sum(idx2)
print "intersection between them : " , sum(np.logical_and(idx1,idx2))

# make feature matrix
# make an empty data frame
rSqDataFrame = pandas.DataFrame( columns=(['bkgSNP'] + list(adniSNPFound)) , index=range(len(bkgSNPFound)) )
dPrDataFrame = pandas.DataFrame( columns=(['bkgSNP'] + list(adniSNPFound)) , index=range(len(bkgSNPFound)) )
lodDataFrame = pandas.DataFrame( columns=(['bkgSNP'] + list(adniSNPFound)) , index=range(len(bkgSNPFound)) )

print "shape of feature data frame : " , rSqDataFrame.shape

# iterate over background SNPs
cnt = 0
for rs in bkgSNPFound:
    rSqDataFrame.bkgSNP[cnt] = bkgSNPFound[cnt]
    dPrDataFrame.bkgSNP[cnt] = bkgSNPFound[cnt]
    lodDataFrame.bkgSNP[cnt] = bkgSNPFound[cnt]
    idx1 = np.logical_and( np.array( hapMapLDAdni['rs1'] == bkgSNPFound[cnt] ) , np.array( hapMapLDAdni['rs2'].isin( adniSNPFound )   ) )
    adniSet1 = np.array( hapMapLDAdni.rs2[ idx1 ])
    idx2 = np.logical_and( np.array( hapMapLDAdni['rs1'].isin( adniSNPFound )  ) , np.array( hapMapLDAdni['rs2'] == bkgSNPFound[cnt]  ) )
    adniSet2 = np.array( hapMapLDAdni.rs1[ idx2 ])
    rSqDataFrame.irow(cnt)[adniSNPFound] = 0
    rSqDataFrame.irow(cnt)[adniSet1] = hapMapLDAdni['Rsquare'][ idx1 ]
    rSqDataFrame.irow(cnt)[adniSet2] = hapMapLDAdni['Rsquare'][ idx2 ]
    dPrDataFrame.irow(cnt)[adniSNPFound] = 0
    dPrDataFrame.irow(cnt)[adniSet1] = hapMapLDAdni['Dprime'][ idx1 ]
    dPrDataFrame.irow(cnt)[adniSet2] = hapMapLDAdni['Dprime'][ idx2 ]
    lodDataFrame.irow(cnt)[adniSNPFound] = 0
    lodDataFrame.irow(cnt)[adniSet1] = hapMapLDAdni['LOD'][ idx1 ]
    lodDataFrame.irow(cnt)[adniSet2] = hapMapLDAdni['LOD'][ idx2 ]
    #print np.array(rSqDataFrame.irow(cnt)[adniSet1] )     
    #print sum(idx1),len(adniSet1),'====',sum(idx2),len(adniSet2)
    #print sum(np.logical_and(idx1 , idx2))
    if np.mod(cnt,100)==0:
        print "cnt : ", cnt
    cnt = cnt + 1
    

# save the results to csv files    
rSqDataFrame.to_csv('/data/vision/polina/users/kayhan/Experiments/DPP/geneticResults/Data/rSquarefeatureTable.csv',index=False)
dPrDataFrame.to_csv('/data/vision/polina/users/kayhan/Experiments/DPP/geneticResults/Data/DPrimefeatureTable.csv',index=False)
lodDataFrame.to_csv('/data/vision/polina/users/kayhan/Experiments/DPP/geneticResults/Data/LODfeatureTable.csv',index=False)
```

# Read adni SNP data and save in MATLAB format

Read the SNPs in:

```python
import numpy as np
from plinkio import plinkfile

adni_bed2_ceu_fn = "/data/vision/polina/projects/ADNI/data/processing/pre-imputations/adni_bed2_ceu_flipped_clean"

plink_file = plinkfile.open( adni_bed2_ceu_fn )
if not plink_file.one_locus_per_row( ):
     print( "This script requires that snps are rows and samples columns." )
     exit( 1 )

sample_list = plink_file.get_samples( )
locus_list = plink_file.get_loci( )

#
#for locus, row in zip( locus_list, plink_file ):
#    for sample, genotype in zip( sample_list, row ):
#        print( "Individual {0} has genotype {1} for snp {2}.".format( sample.iid, genotype, locus.name ) )
```

Convert PLINK data to numpy array

```python
# make sure sample and locus
print "number of subjects in the list ", len(sample_list)
print "number of locus in the list : " , len(locus_list)   

adniSNPData = np.zeros( ( len(locus_list) , len(sample_list)  ) , dtype=np.float)
cnt = 0
for locus, row in zip( locus_list, plink_file ):
    #print (np.array(row)).shape
    adniSNPData[cnt,:] = np.array(row)
    cnt = cnt + 1
 
# some sanity check
AACount = sum(adniSNPData==0)
AaCount = sum(adniSNPData==1)
aaCount = sum(adniSNPData==2)
naCount = sum(adniSNPData==3)
print '# AA :', AACount, ' # (Aa) :', AaCount, ' # (aa) :', aaCount, '# NA :', naCount
print '# (AA) / # (AA) :', 1, ' # (Aa) / # (AA) :', (0.0 + AaCount)/AACount, ' # (aa) / # (AA) :', (0.0 + aaCount)/AACount, '# NA / # (AA):', (0.0 + naCount)/AACount

# visualize part of matrix
imshow(adniSNPData[0:400,0:400])
colorbar()
any(np.isnan(adniSNPData[:]))


```

```python
import numpy, scipy.io

adniSNPData[adniSNPData==3] = np.nan

print "saving the ADNI file ..."
adniMATFn = '/data/vision/polina/projects/ADNI/data/processing/pre-imputations/adni_bed2_ceu_flipped_clean.mat'
scipy.io.savemat(adniMATFn, mdict={'adniSNPData': adniSNPData, 'locus_list':locus_list, 'sample_list':sample_list})

```
